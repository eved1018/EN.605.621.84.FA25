## 1b. English-language explanation

<i> In subsequent assignments, you will have to provide the English-language explanation of your pseudocode; however, for this assignment, the explanation is provided to you </i>

This notebook implements two separate approaches for solving the Fibonacci sequence.  The recursive approach (fib1) simply implements the definition of the Fibonacci sequence, as given in the problem statment.  The base case is defined as F(0), F(1), or F(2).  In these cases, the algorithm simply returns the defined value for F() and reports one step taken.  For i > 2, the program recursively calls fib1(i-1) and fib1(i-2) to be able to define the value of F(i).  At each level of recursion, the program accumulates the number of steps taken by its child processes, and then adds three steps for the workload at its own level. 

In the non-recursive, or bottom-up, implementation (fib2), the algorithm builds a list with each successive Fibonacci number in it, beginning with the values F(0) = 0, F(1) = 1, and F(2) = 1.  Then it iterates up to the given value n, adding the previously calculated values for F(i-1) and F(i-2) and appending the newly-calculated values for F(i) to the accumulating list and adding one step to the accumulating workload counter.  Finally, it returns the sum total of all the work done and the final value in the list of Fibonacci numbers.

## 1c. Code implementation (40 pts)

```python
# CODE CELL

def fib1(n_th_fib_number: int) -> tuple[int, int]:
    """
    This function computes the nth Fibonacci number using recursion.
    Input:
        n_th_fib_number (int) non-negative int determined the Fibonacci number to computer
    Return
        result (int) the nth Fibonacci Number
        steps (int) the number of recursive steps taken 
    
    """
    # thi is a recursive implementation of the Fibonacci sequence
    # returns fibonacci number, steps
    
    # Implement the pseudocode given above for fib1(n)
     
    if n_th_fib_number == 0: # base case: we should return F0 which is defined as 0
        return 0, 1
    
    elif n_th_fib_number == 1 or n_th_fib_number == 2: # base case: we should return F1 and F2 which is defined as 1
        return 1, 1
    
    elif n_th_fib_number > 2:
        # recursive step 
        first_result, first_steps = fib1(n_th_fib_number-1) 
        second_result, second_steps = fib1(n_th_fib_number-2) 
        return first_result + second_result , first_steps + second_steps 
    
    else:
        raise ValueError(f"Input n should be a non-negative int instead got {n_th_fib_number}")

def fib2(n_th_fib_number: int) -> tuple[int, int]:
    """
    This function computes the nth Fibonacci number using iteration.
    Input:
        n_th_fib_number (int) non-negative int determined the Fibonacci number to computer
    Return
        result (int) the nth Fibonacci Number
        steps (int) the number of recursive steps taken 
    
    """
    # this is a non-recursive, bottom-up implementation
    # returns fibonacci number, steps
    
    # Implement the pseudocode given above for fib2(n)
    # Return signature should be:

    if n_th_fib_number == 0: # base case: we should return F0 which is defined as 0
        return 0, 1
    
    elif n_th_fib_number == 1 or n_th_fib_number == 2:
        return 1, 1
    
    elif n_th_fib_number > 2:
        results = [0,1,1]
        steps = 1
        for i in range(2, n_th_fib_number+1):
            results.append(results[i-1] + results[1-2])
            steps += 1 
        return results[n_th_fib_number], steps
    
    else:
        raise ValueError(f"Input n should be a non-negative int instead got {n_th_fib_number}")
```

### Analysis of Recursive Fibonacci Sequence
The recurrence relation for this algorithm is of the form: T(n) = T(n-1) + T(n-2) + f(n). Since it is not of the from T(n) = aT(n/b) + f(n) and the masters theorem cannot be used. 

Since each recursive step does 2 function calls a good guess for the upper bounds of the runtime could be O(2^n). The f(n) component for each recurrence call is factor of a constant time operation (1 addition, 2 subtractions and a comparison).


Using the substitution method outlines in CLRS[1]:

$$ T(n) = T(n-1) + T(n-2) + O(1) = c2^{n-1} + c2^{n-2} + O(1) $$


From exponent rules [trick taken from mod_03_Content.pdf from coursework]:  $$ c2^{n-1} = c2^{n-2} * 2^1 $$ 

$$ c2^{n-1} + c2^{n-2} + O(1) = c2^{n-2} * 2^1 + c2^{n-2} + O(1) = c2^{n-2} * (2 + 1) +  O(1) = 3c2^{n-2} + O(1) \leq c'2^n $$

The induction step requires proving that the inequality holds at a boundery condition. For example we know T(2) has a runtime of 4 ((1 addition, 2 subtractions and a comparison))

$$ T(2) \leq c2^2 = 4c $$

which is true if c=1

This shows that O(2^n) is a valid upper bounds to to fib1().

An interesting finding is that the recurrence relationship T(n) looks like the definition of the Fibonacci sequence with the addition of constant time operations per recurrence. For this reason a guess for a lower bounds of fib1() could be $\Omega$(Fib(n)). One way of verifying this guess is by looking at the graph of the values of fib1(). 



### Analysis of Iterative Fibonacci Sequence
The runtime is of fib2() is O(n) as we loop over the numbers from 2..n and perform 3 constant time operation within the loop, 2 lookups and addition. This means the body of the loop is 3O(1) = O(1).

The resulting runtime is O(n) * O(1) = O(n)
